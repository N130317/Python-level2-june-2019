```python
# Linear graph

import matplotlib.pyplot as plt

plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time

distance = list(range(2,11,2))
time = list(range(1,6,1))
plt.title("Distance vs Time")
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,"bo")
plt.grid()
plt.show()
```

```python
# v = a * t

t = list(range(1,11))
a = list(range(len(t)+1))
v = [a[i] * i for i in t]
plt.plot(t,v,'ro')
plt.grid()
plt.show()
```

```python
n = list(range(10))
sqn = [i**2 for i in n]
plt.plot(n,sqn,'bo')
plt.grid()
plt.show()
```

```python
# Votlage = Time
import numpy as np

t = [i/10 for i in range(1,21)]

v = [1+np.sin(2*np.pi*i) for i in t]
plt.title('Time vs Voltage Graph')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013
import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')
years = df.columns[2:]

incomeArkansas = df.values[3, 2:]

plt.title('Income of Akansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years, incomeArkansas)
plt.grid()
plt.show()
```

```python
# Income of all states from 2005 to 2013
df
incomeallstates = []

for i in df.columns[2:]:
      incomeallstates.append(df[i].mean())
        
incomeallstates
plt.title('Income of all states')
plt.xlabel('Years')
plt.ylabel('Income')
plt.plot(years,incomeallstates)
plt.grid()
plt.show()
```
